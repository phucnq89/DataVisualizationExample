```python
# Import Library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Some configs
plt.rcParams['figure.figsize'] = (12,10)
plt.rcParams['figure.dpi'] = 200
plt.rcParams['font.size'] = 13
sns.set(style='darkgrid') # Thiết lập style cho biểu đồ
import ssl
# Cấu hình ssl cho phép tải dữ liệu mẫu thông qua thư viện
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
# Load sample dataset
# sample_datasets = sns.get_dataset_names()
# print(sample_datasets)
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='sex', size='size')
```

```python
dat = pd.read_csv('./data/Income.csv')
sns.relplot(x='Income', y='Expenditure', data=dat, kind='line')
```

```python
df = pd.read_csv('./data/Housing.csv')
# Randomly select 400 rows from the df
dat = df.sample(400)
sns.scatterplot(x='housing_median_age', y='median_house_value', data=dat, hue='ocean_proximity', size='median_income')
```

```python
df = pd.read_csv('./data/Life_Expectancy.csv')
dat = df[df['Country'].isin(['China', 'Japan', 'Germany', 'France', 'Viet Nam', 'Republic of Korea'])]
sns.scatterplot(x='Life expectancy', y='Schooling', data=dat, hue='Country')
```

```python
df = pd.read_csv('./data/Housing.csv')
# Randomly select 1000 rows from the df
dat = df.sample(1000)
sns.lineplot(x='housing_median_age', y='median_house_value', data=dat, hue='ocean_proximity')
```

```python
df = pd.read_csv('./data/Housing.csv')
# Randomly select 500 rows from the df
dat = df.sample(500)
sns.catplot(x='ocean_proximity', y='median_house_value', data=dat, kind='violin', palette='colorblind')
```

```python
np.random.seed(10) # -> Random với giá trị không đổi
dat = np.random.normal(12, 2, 400)
sns.displot(dat, kde=True, color='darkgreen')
plt.xlabel('Salary')
```

```python
df = pd.read_csv('./data/Income.csv')
sns.regplot(x='Income', y='Expenditure', data=df)
```

```python
df = pd.read_csv('./data/Car_Prices.csv')
sns.lmplot(x='Age(yrs)', y='Sell Price($)', data=df, hue='Manufacturer', markers=['o', 'x'])
```

```python
df = pd.read_csv('./data/Car_Prices.csv')
sns.pairplot(df[['Age(yrs)','Mileage', 'Sell Price($)']])
```

```python
flights_long = sns.load_dataset('flights')
flights = flights_long.pivot('month', 'year', 'passengers')
sns.heatmap(flights, annot=True, fmt='d', linewidths=.5, cmap='YlGnBu')
```
